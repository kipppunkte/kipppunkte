```python
import pandas as pd
import logging
logging.basicConfig(level=logging.INFO)
import textwrap
from pathlib import Path
```

```python
from bs4 import BeautifulSoup
import logging
from pathlib import Path
soup = BeautifulSoup(Path("Kipppunkte.kml").read_text(), 'lxml')
stations = Path("stations.csv").read_text().split("\n")
stations = [_.split("\t", 4) for _ in stations if _.strip()]
s_ids = [_[0] for _ in stations]
placemarks = soup.find_all("placemark")
res = []
coordinates_lut = {}
for pm in placemarks:
    name = pm.find("name").text
    coordinates = pm.find("coordinates").text
    coordinates = list(map(float, coordinates.strip().split(",")[:2]))
    if " " not in name:
        logging.error(pm.find("name").text)
        logging.error(coordinates)
        continue
    s_id, name = name.split(" ", 1)
    s_id = s_id.strip()
    if s_id not in s_ids:
        logging.error(pm.find("name").text)
        continue
        
    name = name.strip()
    coordinates = pm.find("coordinates").text
    coordinates = list(map(float, coordinates.strip().split(",")[:2]))
    coordinates_lut[s_id] = coordinates[::-1]
    print((s_id, *coordinates))
    res.append("\t".join([s_id, str(coordinates[1]), str(coordinates[0])]))
Path("stations_coords.csv").write_text("\n".join(res))
```

```python
fname = Path("./stations_coords.csv")
assert fname.is_file()
s_coords = pd.read_csv(fname, sep="\t", header=None, index_col=0, dtype="str").astype("float").apply(tuple, 1).to_dict()
s_coords = coordinates_lut
s_coords
```

```python

fname = Path("./Stations.xlsx")
assert fname.is_file()
sdf = pd.read_excel(fname, dtype="str").fillna('')
sdf = sdf.map(lambda s: s.strip())
sdf.columns = sdf.columns.str.strip()
sdf["IsGame"] = sdf["IsGame"].str.lower() == "x"
sdf["HasPics"] = sdf["HasPics"].str.lower() == "x"
sdf["HasAudio"] = sdf["HasAudio"].str.lower() == "x"

DOCS_PATH = Path("./docs")
ASSETS_PATH = DOCS_PATH / "assets"
DATA_PATH = Path("GoogleDocs")
PNG_FILES = sorted(DATA_PATH.glob("*.png"))
SITE = 'https://kipppunkte.github.io/kipppunkte'


def create_audio(mp3_name):
    return textwrap.dedent(rf'''
<audio controls>
    <source src="https://github.com/kipppunkte/kipppunkte/raw/gh-pages/assets/{mp3_name}" type="audio/mpeg">
    Your browser does not support the audio tag.
</audio>
''').strip()

def create_game(img_name, img_name1):
    return textwrap.dedent(rf'''
===+ "Auftrag"

    ![Image title](assets/{ img_name }){{: style="max-height:60vh" }}


=== "Ergebnis"

    ![Image title](assets/{ img_name1 }){{: style="max-height:60vh" }}
''').strip()



#print(create_game("name0", "name1"))


def get_directions_url(lon, lat):
    return rf"https://www.google.com/maps/dir/?api=1&travelmode=walking&destination={lon},{lat}"
    
def get_directions_icon(lon, lat, size=24):
    url = get_directions_url(lon, lat)    
    return f'''<a href="{url}"><img src="https://github.com/kipppunkte/kipppunkte/raw/gh-pages/assets/google-maps.svg" width="{size}" height="{size}"></a>'''


def move_to_docs(filename):
    import shutil
    shutil.copy2(filename, ASSETS_PATH / filename.name)

from dataclasses import dataclass

@dataclass
class Station:
    ID:str
    Name:str
    ID:bool
    IsGame:bool
    HasPics:bool
    HasAudio:bool
    Address:str
    Template:str
    DescriptionNextStation:str
    nextStation:Station = None

    @property
    def n_pictures(self):
        if "[[game]]" in self.Template:
            return 2
        return 0

    @property
    def n_audios(self):
        if "[[audio]]" in self.Template:
            return 2
        if "[[game]]" in self.Template:
            return 2
    
    def find_pictures(self):
        return sorted(
            DATA_PATH.glob(f"{self.ID}_*.png")
        )

    def find_audio(self):
        return sorted(
            DATA_PATH.glob(f"{self.ID}_*.mp3")
        )

    def render_template(self):
        logging.info(f"{self.ID} {self.Name}")
        lines = self.Template.strip().split("\n")
        n_audios = sum([1 if _.strip() == "[[audio]]" else 0 for _ in lines])
        audios = self.find_audio()
        pics = self.find_pictures()
        if n_audios:
            assert self.HasAudio, self.ID
        assert n_audios == len(audios), (self.ID, len(audios), n_audios)
        n_pics = sum([2 if _.strip() == "[[game]]" else (1 if _.strip() == "[[picture]]" else 0) for _ in lines])
        if n_pics:
            assert self.HasPics, self.ID
        assert n_pics == len(pics), (self.ID, len(pics), n_pics)

        for i, l in enumerate(lines):
            if l == "[[audio]]":
                a_fname = audios[0]
                move_to_docs(a_fname)
                lines[i] = "\n\n" + create_audio(a_fname.name) + "\n\n"
            elif l == "[[game]]":
                a_fname = pics[0]
                b_fname = pics[1]
                move_to_docs(a_fname)
                move_to_docs(b_fname)
                lines[i] = create_game(a_fname.name, b_fname.name) + "\n\n"
            elif l == "[[picture]]":
                assert len(pics) == 1
                a_fname = pics[0]
                move_to_docs(a_fname)
                lines[i] = f'''![Image title](assets/{ a_fname.name }){{: style="max-height:60vh" }}\n\n'''
            

        current_directions = get_directions_icon(*self.coords)
        
        header = f'''# Station {self.ID}: {self.Name}  {current_directions}\n\n'''
        header += f'''<small>Addresse:<em style="margin-left: 10px">{self.Address}</em></small>\n\n'''
        
        next_directions = ""
        if (self.nextStation is not None) and (self.next_coords is not None):
            next_url = get_directions_url(*self.next_coords)    
            next_d = f'''<a href="{next_url}"><img src="https://github.com/kipppunkte/kipppunkte/raw/gh-pages/assets/google-maps.svg" style="height: 1.5em;margin-right: 0.5em"></a>'''
            next_directions = f'''\n\n____\n\n{next_d}**[Weg zur nächsten Station]({next_url})**\n\n'''
        
        next_description = ""
        if self.DescriptionNextStation:
            if self.nextStation is None:
                next_description = "\n\n____"
            next_description += '''\n\n<img src="https://github.com/kipppunkte/kipppunkte/raw/gh-pages/assets/eyes.svg" style="height: 1.5em;background: white;margin-right: 0.5em">**Halte Ausschau nach:**\n\n'''
            next_description += "<br/>".join(self.DescriptionNextStation.split("\n"))
                
        text = header + "\n\n" + "\n".join(lines) + "\n\n" + next_directions + next_description

        
        return text
        
    @property
    def docs_fname(self):
        return DOCS_PATH / f"{self.ID}.md"
        
    @property
    def coords(self):
        return s_coords[self.ID]
        
    @property
    def next_coords(self):
        if self.nextStation is not None:
            try:
                return self.nextStation.coords
            except:
                pass
                


        

stations = sdf.fillna('').to_dict(orient="records")
stations_d = { sd['ID']: Station(**sd) for sd in stations }
s_ids = list(stations_d.keys())
for i, s_id in enumerate(s_ids):
    sd = stations_d[s_id]
    try:
        sd.nextStation = stations_d[s_ids[i+1]]
    except IndexError:
        next_id = None
    txt = sd.render_template()
    sd.docs_fname.write_text(txt)
```
